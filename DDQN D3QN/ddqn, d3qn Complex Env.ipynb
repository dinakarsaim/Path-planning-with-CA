```python
import csv
import matplotlib.pyplot as graph
import numpy
from env import obstacle_width
from env import Obstacleacc
obcoords = Obstacleacc()
x_o = obcoords.xpart
y_o = obcoords.ypart
with open('/kaggle/working/d3qn_final_path.csv') as file:
    reader  = csv.reader(file)
    for row in reader:
        finalx = numpy.array(list(map(int,row)))
        break
    for row in reader:
        finaly = numpy.array(list(map(int,row)))
        break
graph.figure()
graph.quiver(finalx[:-1],finaly[:-1],finalx[1:]-finalx[:-1],finaly[1:]-finaly[:-1], scale_units='xy', angles='xy', scale=1, color='black')
for i in range(len(x_o)):
    rectangle = graph.Rectangle(( 10* (x_o[i]-0.5), 10*(10 - y_o[i] -0.5)), obstacle_width, obstacle_width, fc='blue',ec="blue")
    graph.gca().add_patch(rectangle)
graph.scatter(10,0, marker = "s", ec = 'k', c ='red', s=100, label ="Start")
graph.scatter(90,100, marker = "s", ec = 'k', c ='red', s =100,label="Target")
graph.grid(linestyle=':')
graph.xlim(0,100)
graph.ylim(0,100)
graph.xlabel('x (m)',size = '14')
graph.ylabel('y (m)',size = '14')
graph.gca().set_aspect('equal', adjustable='box')
graph.show()
```
